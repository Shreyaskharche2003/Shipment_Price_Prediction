```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head(5)
```

```python
df=df[["Unit of Measure (Per Pack)","Line Item Quantity","Pack Price","Unit Price","Freight Cost (USD)","Line Item Insurance (USD)","Shipment Mode","Line Item Value"]]
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.dropna()
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
count = []
for items in df["Freight Cost (USD)"]:
    if(not isfloat(items)):
        # df.drop(items, axis=0)
        count.append(items)
count
```

```python
for items in count:
    # create a Boolean mask for the rows to remove
    mask = df['Freight Cost (USD)'] == items
    
    # select all rows except the ones that contain 'Coca Cola'
    df = df[~mask]
    
    # print the resulting DataFrame
df
```

```python
df["Shipment Mode"] = df["Shipment Mode"].astype(str)
df = pd.get_dummies(df, columns=["Shipment Mode"])
df.head(25)
```

```python
df["Shipment Mode_Air Charter"]=df["Shipment Mode_Air Charter"].astype(int)
df["Shipment Mode_Ocean"]=df["Shipment Mode_Ocean"].astype(int)
df["Shipment Mode_Truck"]=df["Shipment Mode_Truck"].astype(int)
df["Shipment Mode_Air"]=df["Shipment Mode_Air"].astype(int)
```

```python
df.tail(50)
```

```python
X=df.drop(["Line Item Value"],axis=1)
y=df["Line Item Value"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=123,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(max_depth=10, random_state=0)
regr.fit(x_train, y_train)
regr.score(x_train, y_train)
```
